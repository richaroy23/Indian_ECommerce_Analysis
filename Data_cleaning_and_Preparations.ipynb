```python
import pandas as pd
import numpy as np
```

```python

customers = pd.read_csv(r"indian_ecommerce_dataset\customers.csv")
order_items = pd.read_csv(r"indian_ecommerce_dataset\order_items.csv")
orders = pd.read_csv(r"indian_ecommerce_dataset\orders.csv")
payments = pd.read_csv(r"indian_ecommerce_dataset\payments.csv")
products = pd.read_csv(r"indian_ecommerce_dataset\products.csv")
sellers = pd.read_csv(r"indian_ecommerce_dataset\sellers.csv")


data = {
    "Customers": customers,
    "Order Items": order_items,
    "Orders": orders,
    "Payments": payments,
    "Products": products,
    "Sellers": sellers
}
```

```python
# ── STEP 1: Drop Duplicates ──────────────────────────────────────────────
print("=== BEFORE Deduplication ===")
for name, df in data.items():
    print(f"{name}: {df.duplicated().sum()} duplicates")

# Drop exact duplicates, keep first occurrence
for name in data:
    before = len(data[name])
    data[name] = data[name].drop_duplicates().reset_index(drop=True)
    after = len(data[name])
    if before != after:
        print(f"  ✓ {name}: dropped {before - after} duplicate rows")

print("\n=== AFTER Deduplication ===")
for name, df in data.items():
    print(f"{name}: {len(df)} rows")
```

```python
# ── STEP 2: Fill Nulls – Customers ────────────────────────────────────────
customers = data['Customers']

customers['customer_email']   = customers['customer_email'].fillna('not_provided')
customers['customer_phone']   = customers['customer_phone'].fillna('not_provided')
customers['customer_pincode'] = customers['customer_pincode'].fillna('000000')

data['Customers'] = customers

print("Customers nulls after fill:")
print(customers.isnull().sum())
```

```python
# ── STEP 3: Fill Nulls – Orders (date columns) ───────────────────────────
orders = data['Orders']

# Convert to datetime so NaT is handled properly
date_cols = ['order_purchase_timestamp', 'order_approved_at',
             'order_delivered_timestamp', 'order_estimated_delivery']
for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')

# ⚠️ Do NOT drop rows with null dates — they are valid pending/undelivered orders
# We keep NaT and let analysis filter them as needed.
print("Orders date nulls (NaT kept intentionally):")
print(orders[date_cols].isnull().sum())

data['Orders'] = orders
```

```python
# ── STEP 4: Fill Nulls – Payments ────────────────────────────────────────
payments = data['Payments']

payments['payment_value']        = payments['payment_value'].fillna(0.0)   # failed transaction = 0
payments['payment_type']         = payments['payment_type'].fillna('Unknown')
payments['payment_installments'] = payments['payment_installments'].fillna(1).astype(int)  # default: 1 installment

data['Payments'] = payments

print("Payments nulls after fill:")
print(payments.isnull().sum())
```

```python
# ── STEP 5: Fill Nulls – Products ────────────────────────────────────────
products = data['Products']

products['brand']        = products['brand'].fillna('Unbranded')
products['weight_grams'] = products['weight_grams'].fillna(products['weight_grams'].median()).astype(int)
products['stock_qty']    = products['stock_qty'].fillna(0).astype(int)  # assume out of stock

data['Products'] = products

print("Products nulls after fill:")
print(products.isnull().sum())
```

```python
# ── STEP 6: Fill Nulls – Sellers ─────────────────────────────────────────
sellers = data['Sellers']

sellers['seller_pincode'] = sellers['seller_pincode'].fillna('000000')
sellers['seller_rating']  = sellers['seller_rating'].fillna(sellers['seller_rating'].median()).round(1)

data['Sellers'] = sellers

print("Sellers nulls after fill:")
print(sellers.isnull().sum())
```

```python
# ── STEP 7: Final Verification ──────────────────────────────────────────────
print("========================================")
print("     FINAL DATA QUALITY SUMMARY")
print("========================================")
for name, df in data.items():
    total_nulls = df.isnull().sum().sum()
    # Exclude NaT date columns from the 'problem' count for Orders
    print(f"\n{name} ({len(df):,} rows)")
    print(f"  Remaining nulls  : {total_nulls}")
    print(f"  Duplicates       : {df.duplicated().sum()}")
    if total_nulls > 0:
        null_cols = df.isnull().sum()
        print(f"  Null columns     : {null_cols[null_cols > 0].to_dict()}")
print("\n✓ Cleaning complete!")
```

```python
# STEP 9: Outlier Detection & Decision
import numpy as np

print('=' * 60)
print('         OUTLIER DETECTION REPORT')
print('=' * 60)

# --- 1. Orders per Customer (IQR method, 3x fence) ---
orders_per_cust = orders['customer_id'].value_counts()
Q1_o = orders_per_cust.quantile(0.25)
Q3_o = orders_per_cust.quantile(0.75)
IQR_o = Q3_o - Q1_o
upper_fence_o = Q3_o + 3 * IQR_o
outlier_custs = orders_per_cust[orders_per_cust > upper_fence_o]

print('\n[1] Orders per Customer')
print('  Median orders      :', int(orders_per_cust.median()))
print('  IQR upper fence    :', int(upper_fence_o))
print('  Outlier customers  :', len(outlier_custs))
print('  Max orders (1 cust):', int(orders_per_cust.max()))
if len(outlier_custs) > 0:
    print('  Top offenders:')
    for cid, cnt in outlier_custs.head(5).items():
        print('   ', cid, '->', cnt, 'orders')

# --- 2. Payment Value ---
pv = payments['payment_value']
Q1_p = pv.quantile(0.25)
Q3_p = pv.quantile(0.75)
IQR_p = Q3_p - Q1_p
upper_fence_p = Q3_p + 3 * IQR_p
outlier_payments = payments[payments['payment_value'] > upper_fence_p]
print('\n[2] Payment Value (Rs.)')
print('  Median payment     : Rs.', round(pv.median(), 2))
print('  IQR upper fence    : Rs.', round(upper_fence_p, 2))
print('  Outlier rows       :', len(outlier_payments))
print('  Max payment        : Rs.', round(pv.max(), 2))

# --- 3. Product Unit Price ---
products_df = data['Products']
pp = products_df['unit_price']
Q1_pp = pp.quantile(0.25)
Q3_pp = pp.quantile(0.75)
IQR_pp = Q3_pp - Q1_pp
upper_fence_pp = Q3_pp + 3 * IQR_pp
outlier_prices = products_df[products_df['unit_price'] > upper_fence_pp]
print('\n[3] Product Unit Price (Rs.)')
print('  Median price       : Rs.', round(pp.median(), 2))
print('  IQR upper fence    : Rs.', round(upper_fence_pp, 2))
print('  Outlier products   :', len(outlier_prices))
print('  Max price          : Rs.', round(pp.max(), 2))

# ── DECISION ──────────────────────────────────────────────
print('\n' + '=' * 60)
print('DECISION SUMMARY')
print('=' * 60)

# DROP: customers with abnormally high order counts (bot/test data)
if len(outlier_custs) > 0:
    valid_custs = orders_per_cust[orders_per_cust <= upper_fence_o].index
    before = len(data['Orders'])
    data['Orders'] = data['Orders'][data['Orders']['customer_id'].isin(valid_custs)].reset_index(drop=True)
    dropped = before - len(data['Orders'])
    print('  Orders/Customer : DROPPED', dropped, 'rows (', len(outlier_custs), 'outlier customers removed)')
else:
    print('  Orders/Customer : No outliers found')

# KEEP: payment and price outliers - valid high-value transactions
print('  Payment values  : KEPT (valid high-value orders)')
print('  Product prices  : KEPT (premium products are legitimate)')
print('\nUpdated Orders shape:', data['Orders'].shape)
```

```python
# STEP 8: Export All Cleaned DataFrames to CSV
import os

output_dir = 'cleaned_data'
os.makedirs(output_dir, exist_ok=True)

export_map = {
    'customers_clean.csv':   data['Customers'],
    'order_items_clean.csv': data['Order Items'],
    'orders_clean.csv':      data['Orders'],
    'payments_clean.csv':    data['Payments'],
    'products_clean.csv':    data['Products'],
    'sellers_clean.csv':     data['Sellers'],
}

print('Saving cleaned CSVs to:', os.path.abspath(output_dir))
for filename, df in export_map.items():
    path = os.path.join(output_dir, filename)
    df.to_csv(path, index=False)
    print('  Saved:', filename, '|', len(df), 'rows |', len(df.columns), 'cols')

print('All 6 cleaned files saved successfully!')
```
